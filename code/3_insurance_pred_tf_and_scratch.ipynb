```python
def log_loss(y_true, y_predicted):
    epsilon = 1e-15
    y_predicted_new = [max(i,epsilon) for i in y_predicted]
    y_predicted_new = [min(i,1-epsilon) for i in y_predicted_new]
    y_predicted_new = np.array(y_predicted_new)
    return -np.mean(y_true*np.log(y_predicted_new) + (1-y_true)*np.log(1-y_predicted_new))
```

```python
def sigmoid_numpy(X):
    return 1/(1+np.exp(-X))

sigmoid_numpy(np.array([12,0,1]))
```

```python
def gradient_descent(age,affordibility,y_true,epochs, loss_thres):
    # returns w1, w2, bias
    w1 = w2 = 1
    bias = 0
    rate = 0.5
    n = len(age)
    
    for i in range(epochs):
        weighted_sum = w1*age + w2*affordibility + bias
        y_predicted = sigmoid_numpy(weighted_sum)
        loss = log_loss(y_true,y_predicted)
        w1 = w1 - (1/n)*rate*(np.dot(np.transpose(age),(y_predicted-y_true)))
        w2 = w2 - (1/n)*rate*(np.dot(np.transpose(affordibility),(y_predicted-y_true)))
        bias = bias - (1/n)*rate*np.mean(y_predicted-y_true)
        print(f'Epoch:{i}, w1:{w1}, w2:{w2}, bias:{bias}, loss:{loss}')
        if loss <=loss_thres:
            break
    return w1,w2,bias
```

```python
class myNN:
    def __init__(self):
        self.w1 = 1
        self.w2 = 1
        self.bias = 0
    
    def gradient_descent(self,age,affordibility,y_true,epochs, loss_thres):
        # returns w1, w2, bias
        w1 = w2 = 1
        bias = 0
        rate = 0.5
        n = len(age)

        for i in range(epochs):
            weighted_sum = w1*age + w2*affordibility + bias
            y_predicted = sigmoid_numpy(weighted_sum)
            loss = log_loss(y_true,y_predicted)
            w1 = w1 - (1/n)*rate*(np.dot(np.transpose(age),(y_predicted-y_true)))
            w2 = w2 - (1/n)*rate*(np.dot(np.transpose(affordibility),(y_predicted-y_true)))
            bias = bias - (1/n)*rate*np.mean(y_predicted-y_true)
            
            if i%50 ==0:
                print(f'Epoch:{i}, w1:{w1}, w2:{w2}, bias:{bias}, loss:{loss}')
            if loss <=loss_thres:
                print(f'Epoch:{i}, w1:{w1}, w2:{w2}, bias:{bias}, loss:{loss}')
                break
        return w1,w2,bias
    
    def fit(self, X, y, epochs, loss_thres):
        self.w1, self.w2, self.bias = self.gradient_descent(X['age'], X['affordibility'], y, epochs, loss_thres)
        
    def predict(self,X_test):
        weighted_sum = self.w1*X_test['age'] + self.w2*X_test['affordibility'] + self.bias
        return sigmoid_numpy(weighted_sum)
```

```python
customModel = myNN()
customModel.fit(X_train, y_train, epochs=5000,loss_thres = 0.4363)
```
